```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Pothole_Image_Dataset/109.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.blur(gray,(3,3))
img_log = (np.log(blur+1)/(np.log(1+np.max(blur))))*255
img_log = np.array(img_log,dtype=np.uint8)
bilateral = cv2.bilateralFilter(img_log, 5, 75, 75)
edges = cv2.Canny(bilateral,100,200)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)
orb = cv2.ORB_create(nfeatures=1500)
keypoints, descriptors = orb.detectAndCompute(closing, None)
featuredImg = cv2.drawKeypoints(closing, keypoints, None)
```

```python
cv2.imwrite('Output/Outcome-109.jpg', featuredImg)
```

```python
import os
input_folder = 'Pothole_Image_Datset'
output_folder = 'Output'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
input_files = os.listdir(input_folder)
for input_file in input_files:
    input_path = os.path.join(input_folder, input_file)
    img = cv2.imread(input_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.blur(gray, (3, 3))
    max_value = np.max(blur)
    img_log = (np.log(blur + 1) / (np.log(1 + max_value + 1e-10))) * 255
    img_log = np.array(img_log, dtype=np.uint8)
    bilateral = cv2.bilateralFilter(img_log, 5, 75, 75)
    edges = cv2.Canny(bilateral, 100, 200)
    kernel = np.ones((5, 5), np.uint8)
    closing = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)
    orb = cv2.ORB_create(nfeatures=1500)
    keypoints, descriptors = orb.detectAndCompute(closing, None)
    featuredImg = cv2.drawKeypoints(closing, keypoints, None)
    output_path = os.path.join(output_folder, f'Outcome-{os.path.splitext(input_file)[0]}.jpg')
    cv2.imwrite(output_path, featuredImg)
    plt.figure(figsize=(15, 5))
    num_images = len(input_files)
    plt.figure(figsize=(10, 2 * num_images))
for i, (input_file, output_file) in enumerate(zip(input_files, os.listdir(output_folder))):
    input_path = os.path.join(input_folder, input_file)
    output_path = os.path.join(output_folder, output_file)
    plt.subplot(num_images, 2, 2 * i + 1)
    plt.imshow(cv2.imread(input_path))
    plt.title(f'')
    plt.xticks([]), plt.yticks([])
    plt.subplot(num_images, 2, 2 * i + 2)
    plt.imshow(cv2.imread(output_path), cmap='gray')
    plt.title(f'')
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```
